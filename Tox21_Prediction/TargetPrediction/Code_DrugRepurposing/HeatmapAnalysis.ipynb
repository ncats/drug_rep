```python
import pandas as pd
```

```python
knn = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/knnModelsForSingleGenes.csv", index_col=0)
SVC = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/SVCModelsForSingleGenes.csv", index_col=0)
xgb = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/xgbModelsForSingleGenes.csv", index_col=0)
rf = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/rfModelsForSingleGenes.csv", index_col=0)
```

```python
knn.head(5)
```

```python
knn['mean'] = knn.mean(axis=1)
knn.head(5)
```

```python
rf['mean'] = rf.mean(axis=1)
xgb['mean'] = xgb.mean(axis=1)
SVC['mean'] = SVC.mean(axis=1)
```

```python
knn = knn[["mean"]]
xgb = xgb[["mean"]]
rf = rf[["mean"]]
SVC = SVC[["mean"]]
```

```python
result = pd.concat([knn, SVC, rf, xgb], axis=1).reindex(xgb.index)
```

```python
result.columns = ['KNN', 'SVC', 'RF', 'XGB']
```

```python
result
```

```python
def add(a, b, c, d):
    return a + b + c + d

result['add'] = result.apply(lambda row: add(row['KNN'],
                                         row['SVC'], row['RF'], row['RF']), axis=1)
 
print('\nAfter Applying Function: ')
# printing the new dataframe
result = result.sort_values(by=['add'],ascending=False)
result
```

```python
result = result.drop(columns=['add'])
```

```python
knn = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/knnModelsForSingleGenes.csv", index_col=0)
SVC = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/SVCModelsForSingleGenes.csv", index_col=0)
xgb = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/xgbModelsForSingleGenes.csv", index_col=0)
rf = pd.read_csv("C:/Users/xueb2/Desktop/vs code/Models/SingleGeneResults/rfModelsForSingleGenes.csv", index_col=0)
```

```python
params = knn.mean(axis=0)
knn_params = params.to_frame().reset_index()
knn_params.columns = ['params', 'roc_auc']

params = SVC.mean(axis=0)
SVC_params = params.to_frame().reset_index()
SVC_params.columns = ['params', 'roc_auc']

params = xgb.mean(axis=0)
xgb_params = params.to_frame().reset_index()
xgb_params.columns = ['params', 'roc_auc']

params = rf.mean(axis=0)
rf_params = params.to_frame().reset_index()
rf_params.columns = ['params', 'roc_auc']
```

```python
knn_best = knn_params[knn_params['roc_auc']==max(knn_params['roc_auc'])]
SVC_best = SVC_params[SVC_params['roc_auc']==max(SVC_params['roc_auc'])]
xgb_best = xgb_params[xgb_params['roc_auc']==max(xgb_params['roc_auc'])]
rf_best = rf_params[rf_params['roc_auc']==max(rf_params['roc_auc'])]
```

```python
best_params = pd.concat([knn_best, SVC_best,xgb_best,rf_best], axis=0).reindex()
best_params
```

```python
best_params.to_csv('best_params.csv')
```
